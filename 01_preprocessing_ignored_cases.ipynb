```python
import gc

def proportion_of_unknowed_rows(df, threshold=0.5):
    """
    Calculates the proportion of rows in a DataFrame where more than a certain
    percentage of columns have the value 9, and removes those rows from the DataFrame.

    Args:
    df: The input DataFrame (modified in place).
    threshold: The proportion of columns that must have the value 9.

    Returns:
    The proportion of rows meeting the criteria before removal.
    """
    gc.collect()
    num_cols = len(df.columns)
    count_9_threshold = int(num_cols * threshold)

    # Identify rows to drop
    rows_to_drop = df.index[(df == 9).sum(axis=1) >= count_9_threshold]

    # Calculate the proportion of rows meeting the criteria
    proportion = len(rows_to_drop) / len(df) if len(df) > 0 else 0

    # Drop rows in place
    df.drop(index=rows_to_drop, inplace=True)

    return proportion
```
